```python
numeric_columns = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights',
                   'adults', 'children', 'babies', 'previous_cancellations',
                   'previous_bookings_not_canceled', 'booking_changes',
                   'days_in_waiting_list', 'adr', 'required_car_parking_spaces',
                   'total_of_special_requests']

# Function to remove outliers based on the IQR method
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers for each numeric column
for column in numeric_columns:
    df = remove_outliers(df, column)
```

```python
plt.figure(figsize=(16, 12))
for column in numeric_columns:
    plt.subplot(4, 4, numeric_columns.index(column) + 1)
    sns.boxplot(data=df, y=column)
    plt.title(f'Boxplot for {column}')
plt.tight_layout()
plt.show()
```

### DATA ANALYSIS AND VISUALIZATION

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
cancelled_perc
# Data
labels = ['Not Canceled', 'Canceled']
sizes = cancelled_perc.values
# Colors
colors = ['cyan', 'magenta']
explode = (0, 0.1)

# Create a pie chart
plt.figure(figsize=(5, 5))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Cancellation Percentage')
plt.axis('equal')

plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1= sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
cancellation_perc = resort_hotel['is_canceled'].value_counts(normalize=True)
cancellation_perc
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
cancellation_perc = city_hotel['is_canceled'].value_counts(normalize=True)
cancellation_perc
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Avg Daily Rate in City and Resort Hotel', fontsize = 20)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (8,4))
ax1= sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation status per month', size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

```python
canceled_reservations = df[df['is_canceled'] == 1]

plt.figure(figsize=(20, 8))
plt.title('Avg Daily Rate per month', fontsize=20)
sns.barplot(x='month', y='adr', data=canceled_reservations, estimator=sum)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
canceled_reservations = canceled_reservations.groupby('reservation_status_date')[['adr']].mean()
canceled_reservations.reset_index(inplace = True)
canceled_reservations.sort_values('reservation_status_date', inplace = True)

not_canceled_reservations = df[df['is_canceled'] == 0]
not_canceled_reservations_adr = not_canceled_reservations.groupby('reservation_status_date')[['adr']].mean()
not_canceled_reservations_adr.reset_index(inplace = True)
not_canceled_reservations_adr.sort_values('reservation_status_date', inplace = True)
```

```python
plt.figure(figsize=(20, 6))
sns.set(style='whitegrid')

sns.lineplot(x='reservation_status_date', y='adr', data=canceled_reservations, label='Canceled', marker='o')
sns.lineplot(x='reservation_status_date', y='adr', data=not_canceled_reservations, label='Not Canceled', marker='o')

# Set labels and title
plt.title('Average Daily Rate (ADR) Over Time', fontsize=16)
plt.xlabel('Reservation Status Date', fontsize=12)
plt.ylabel('ADR', fontsize=12)
plt.xticks(rotation=45)
plt.legend(fontsize=12)

plt.show()
```

```python
# Convert 'reservation_status_date' column to datetime if it's not already
canceled_reservations['reservation_status_date'] = pd.to_datetime(canceled_reservations['reservation_status_date'])
not_canceled_reservations['reservation_status_date'] = pd.to_datetime(not_canceled_reservations['reservation_status_date'])

# Filter for dates within the specified range
canceled_reservations = canceled_reservations[
    (canceled_reservations['reservation_status_date'] > '2016-01-01') &
    (canceled_reservations['reservation_status_date'] < '2017-09-01')
]

not_canceled_reservations = not_canceled_reservations[
    (not_canceled_reservations['reservation_status_date'] > '2016-01-01') &
    (not_canceled_reservations['reservation_status_date'] < '2017-09-01')
]
```

```python
# Set the figure size and style
plt.figure(figsize=(16, 8))
sns.set(style='whitegrid')

# Create a line chart for canceled reservations
ax1 = sns.lineplot(x='reservation_status_date', y='adr', data=canceled_reservations, label='Canceled', marker='o')
ax1.set_xlabel('Reservation Status Date')
ax1.set_ylabel('ADR for Canceled Reservations', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.set_title('ADR Over Time for Canceled Reservations', fontsize=16)
ax1.legend(loc='upper left', fontsize=12)

# Create a secondary y-axis for not canceled reservations
ax2 = ax1.twinx()
sns.lineplot(x='reservation_status_date', y='adr', data=not_canceled_reservations, label='Not Canceled', marker='o', color='tab:orange')
ax2.set_ylabel('ADR for Not Canceled Reservations', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')
ax2.legend(loc='upper right', fontsize=12)

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
